```python
def pairplot_func(to_compare,filename):
    sns.set(font_scale=1.5)
    to_compare.extend(['species'])
    fig = sns.pairplot(data = df[to_compare], hue='species', corner = 'true')
    fig._legend.remove()
    fig.savefig(f"./experiments/prop_correlations_manuscript/{filename}.png")
    plt.close()
```

```python
plot_list = [['lattice_constant_fcc','re','rc'],
             ['bulk_modulus_fcc','Ec','b0'],
             ['c44_fcc','rho0','zbl'],
             ['c44_fcc','bulk_modulus_fcc','c11_fcc'],
             ['cohesive_energy_fcc','Ec','esub'],
             ['surface_energy_100_fcc', 'Ec','esub']
]
filenames = ['lattice_constant_fcc',
             'bulk_modulus_fcc',
             'c44_fcc',
             'c44_fcc_2',
             'cohesive_energy_fcc',
             'surface_energy_100_fcc']

for current,filename in zip(plot_list,filenames):
    pairplot_func(current,filename)
```
